```python
def get_with_tsne(words, a, b, df):
    axis = df.loc[b] - df.loc[a]
    df_words = df.loc[words]
    df_words_rest = df_words - pd.DataFrame(np.outer(df_words.dot(axis), axis), index=df_words.index, columns=df_words.columns)
    model = TSNE(n_components=1, random_state=0, perplexity=100)
    df_words_tsne = pd.DataFrame(model.fit_transform(df_words_rest), index=words).rename(columns={0:"tsne"})
    df_words_tsne["name"]=words
    df_words_tsne["axis"] = df_words.dot(axis)
    return df_words_tsne
```

```python
adjectives_tsne = get_with_tsne(["long", "longer", "short", "shorter", "dark", "darker", "strong", "stronger", "soft", "softer"]\
                                , "short", "shorter")
adjs = ["long", "longer", "short", "shorter", "dark", "darker", "strong", "stronger", "soft", "softer"]
```

```python
%%R -i adjectives_tsne

ggplot(adjectives_tsne, aes(tsne, axis, label=name)) +
  geom_text()
```

```python
def get_with_axes(words, a, b, c, d, df):
    ab = df.loc[a] - df.loc[b]
    cd = df.loc[c] - df.loc[d]
    proj = pd.DataFrame([ab, cd], index=["ab", "cd"]).transpose()
    df_plot = df.loc[words].dot(proj)
    df_plot["name"] = df_plot.index
    return df_plot
```

```python
adjs_plot = get_with_axes(adjs, "short", "shorter", "more", "less")
```

```python
%%R -i adjs_plot

ggplot(adjs_plot, aes(ab, cd, label=name)) +
  geom_text()
```

```python
gender_words=["man", "woman", "he", "she", "girl", "boy", "waiter", "waitress", "uncle", "aunt", "mother", "father", "actor", "actress"]
gender_plot = get_with_axes(gender_words, "he", "she", "small", "big")
```

```python
%%R -i gender_plot

ggplot(gender_plot, aes(ab, cd, label=name)) +
  geom_text()
```
